```python
from tbparse import SummaryReader
```

```python
log_dir = "../lightning_logs/"
```

```python
reader = SummaryReader(log_dir, extra_columns={'dir_name', 'file_name'}, pivot=True)
metrics_df = reader.scalars
```

```python
hyperparams_df = reader.hparams
```

```python

wav2vec_df_hyperparams = hyperparams_df[hyperparams_df['dir_name'].map(lambda x: "wav2vec" in x.split('/')[0])].copy()
wav2vec_df_metrics = metrics_df[metrics_df['dir_name'].map(lambda x: "wav2vec" in x.split('/')[0])].copy()

wav2vec_df_metrics = wav2vec_df_metrics[['dir_name', 'test_acc']].copy()
wav2vec_df_metrics.set_index('dir_name', inplace=True)


wav2vec_df_hyperparams = wav2vec_df_hyperparams.drop(columns=['file_name']).drop_duplicates()
wav2vec_df_hyperparams.set_index('dir_name', inplace=True)

# preserve only those that finished training
wav2vec_df_metrics = wav2vec_df_metrics.dropna()
```

```python
wav2vec_df = wav2vec_df_metrics.join(wav2vec_df_hyperparams, how='inner')
wav2vec_df
```

```python
wav2vec_df['trimmed'] = wav2vec_df.index.str.contains('notrim')
wav2vec_df['model_type'] = wav2vec_df.index.str.split('_').str[1]

wav2vec_df
```

```python
# flip trimmed column
wav2vec_df['trimmed'] = wav2vec_df['trimmed'].map(lambda x: not x)
```

```python
mean_results = wav2vec_df.groupby(['model_type', 'trimmed', 'num_layers', 'num_classes', 'input_size', 'hidden_size', 'classes']).agg({'test_acc':['mean', 'std']})
# drop classes column
mean_results.reset_index(inplace=True)
mean_results.columns = mean_results.columns.droplevel(1)
mean_results.drop(columns=['classes', 'input_size'], inplace=True)``
# # sort by test_acc
mean_results.columns = ['model_type', 'trimmed', 'num_layers', 'num_classes', 'hidden_size', 'acc mean', 'acc std']
mean_results.sort_values(by='acc mean', ascending=False, inplace=True)
# # rename columns
# round acc to 3 decimals
mean_results['acc mean'] = mean_results['acc mean'].map(lambda x: round(x, 3))
mean_results['acc std'] = mean_results['acc std'].map(lambda x: round(x, 5))
# drop num_classes column
mean_results.drop(columns=['num_classes'], inplace=True)
# set num_layers, hidden_size to int
mean_results['num_layers'] = mean_results['num_layers'].map(lambda x: int(x))
mean_results['hidden_size'] = mean_results['hidden_size'].map(lambda x: int(x))
mean_results
```
